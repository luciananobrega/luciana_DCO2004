## Plot do espectro dos arquivos Violino.wav e Flauta.wav

### 1. Matlab

```python
close all;clc;clear all;

% Abrindo os arquivos de áudio (vt: amplitude das amostras, fs: frequência
% de amostragem
[vtViolino, fsViolino] = audioread('./Material/Violino.wav');
[vtFlauta, fsFlauta] = audioread('./Material/Flauta.wav');

% Figura 1: Espectro de Violino.wav
subplot(2,1,1)
fViolino = linspace(-fsViolino/2, fsViolino/2, size(vtViolino,1));
FViolino = fftshift(abs(fft(vtViolino)))/length(fViolino);
plot(fViolino, FViolino);
axis([0 1e4 0 0.04])
title('Espectro de Violino.wav');
xlabel('Frequência(Hz)');
ylabel('|Amplitude|');

% Figura 2: Espectro de Flauta.wav
subplot(2,1,2)
fFlauta = linspace(-fsFlauta/2, fsFlauta/2, size(vtFlauta,1));
FFlauta = fftshift(abs(fft(vtFlauta)))/length(fFlauta);
plot(fFlauta, FFlauta);
axis([0 1e4 0 0.055])
title('Espectro de Flauta.wav');
xlabel('Frequência(Hz)');
ylabel('|Amplitude|');
```

### 2. Discussão dos resultados

Nas imagens, observa-se o espectro de áudios de dois instrumentos musicais diferentes. Percebe-se que ambos os espectros são compostos pela mesma frequência principal e suas harmônicas, o que indica que os instrumentos estão tocando a mesma nota.

Porém, instrumentos musicais diferentes possuem timbres (formas de onda no tempo) diferentes, resultando em pulsos distintos também na frequência.

### 3. Python

```python
import numpy as np
from numpy.fft import *
import scipy.io.wavfile as wv 
import matplotlib.pyplot as plt;
%matplotlib notebook

# Abrindo arquivos
fsFlauta,vtFlauta = wv.read('./Material/Flauta.wav')
fsViolino,vtViolino = wv.read('./Material/Violino.wav')

# Figura 1: Espectro de Violino.wav
plt.figure(1)
fViolino = np.linspace(-fsViolino/2, fsViolino/2, np.size(vtViolino))
FViolino = fftshift(abs(fft(vtViolino)))/np.size(fViolino)
plt.plot(fViolino, FViolino)
plt.xlim([0,1e4])
plt.title('Espectro de Violino.wav')
plt.xlabel('Frequência(Hz)')
plt.ylabel('|Amplitude|')

# Figura 2: Espectro de Flauta.wav
plt.figure(2)
fFlauta = np.linspace(-fsFlauta/2, fsFlauta/2, np.size(vtFlauta))
FFlauta = fftshift(abs(fft(vtFlauta)))/np.size(fFlauta)
plt.plot(fFlauta, FFlauta)
plt.xlim([0, 1e4])
plt.title('Espectro de Flauta.wav')
plt.xlabel('Frequência(Hz)')
plt.ylabel('|Amplitude|')
```
