### Reproduzindo o código principal em Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros
n_bits = int(1000)          # Número de bits
T = int(500)                # Tempo de símbolo OFDM
Ts = 2                      # Tempo de símbolo em portadora única
K = int(T/Ts)               # Número de subportadoras independentes
N = int(2*K)                # DFT de N pontos
sigmas=[0, 0.1, 1]          # Vetor de variâncias do ruído

# Gerar bits aleatórios
dataIn=np.random.rand(n_bits);   # Sequência de números entre 0 e 1 uniformemente distribuídos
dataIn=np.sign(dataIn-.5);       # Sequência de -1 e 1
# Conversor serial paralelo
dataInMatrix = dataIn.reshape((int(n_bits/4),4))

# Gerar constelaçao 16-QAM
#seq16qam = 2*dataInMatrix[:,0]+dataInMatrix[:,1]+1j*(2*dataInMatrix[:,2]+dataInMatrix[:,3])
seq16 = 2*dataInMatrix[:,0]+dataInMatrix[:,1]+1j*(2*dataInMatrix[:,2]+dataInMatrix[:,3])
#seq16=seq16qam';
# Garantir propriedadade da simetria
X = np.concatenate([seq16, np.conj(seq16[::-1])])

# Construindo xn
xn = np.zeros((N), dtype=complex)
for n in range(N):
    for k in range(N):
        xn[n] = xn[n] + 1/np.sqrt(N)*X[k]*np.exp(1j*2*np.pi*n*k/N)

# Loop de variâncias
for ik in range(len(sigmas)):
    
    # Adição de ruído
    variance = sigmas[ik];
    noise = np.sqrt(variance)*np.random.randn(N)+1j*np.sqrt(variance)*np.random.randn(N)
    # sinal recebido = xn + ruído 
    rn = xn+noise;
    # DFT de rn
    Y = np.zeros((K), dtype=complex)
    for k in range(K):
        for n in range(N):
            Y[k] = Y[k] + 1/np.sqrt(N)*rn[n]*np.exp(-1j*2*np.pi*k*n/N)

    # Plots
    plt.scatter(np.real(Y),np.imag(Y))
    plt.scatter(np.real(seq16),np.imag(seq16))
    
    plt.title('Sinal com ruído de variância ' + str(variance));
    plt.show()
    # Demodulação 
    Z = np.zeros((len(Y)), dtype=complex)
    for k in range(len(Y)): # Para percorrer todo o vetor Yk 
        if np.real(Y[k]) > 0: # Para parte real de Yk positiva
            if np.real(Y[k]) > 2:
                Z[k] = 3
            else:
                Z[k] = 1
        else: # Para parte real de Yk negativa ou igual a zero
            if np.real(Y[k]) < -2:
                 Z[k] = -3
            else:
                 Z[k] = -1

        if np.imag(Y[k]) > 0: # Para parte imaginaria de Yk positiva
            if np.imag(Y[k]) > 2:
                Z[k] = Z[k] + 1j*3
            else:
                Z[k] = Z[k] + 1j

        else: # Para parte imaginaria de Yk negativa ou igual a zero
            if np.imag(Y[k]) < -2:
                 Z[k] = Z[k] - 1j*3
            else:
                 Z[k] = Z[k] - 1j
                    
    # Contagem de erro
    error = len(np.where(Z-X[:250]!=0)[0])
    print('Para variância de ' + str(variance) + ' houve ' + str(error) + ' símbolos errados.')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc

# Parâmetros
n_bits = int(1e6)          # Número de bits
Ts = 2                      # Tempo de símbolo em portadora única
T = int(n_bits/Ts)                # Tempo de símbolo OFDM
K = int(T/Ts)               # Número de subportadoras independentes
N = int(2*K)                # DFT de N pontos
vtEbN0_dB = np.arange(0,15) # Eb/N0 em dB
vtEbN0=10**(vtEbN0_dB/10)   # Eb/N0 linear

# Gerar bits aleatórios
dataIn=np.random.rand(n_bits)   # Sequência de números entre 0 e 1 uniformemente distribuídos
dataIn=np.sign(dataIn-.5)       # Sequência de -1 e 1
# Conversor serial paralelo
dataInMatrix = dataIn.reshape((int(n_bits/4),4))

# Gerar constelaçao 16-QAM
seq16 = 2*dataInMatrix[:,0]+dataInMatrix[:,1]+1j*(2*dataInMatrix[:,2]+dataInMatrix[:,3])

# Garantir propriedadade da simetria
X = np.concatenate([seq16, np.conj(seq16[::-1])])

# Construindo xn
xn = np.zeros((N), dtype=complex)
#xn = (N/np.sqrt(N))*np.fft.ifft(X)
xn = np.fft.ifft(X, norm='ortho')
#for n in range(N):
#    for k in range(N):
#        xn[n] = xn[n] + 1/np.sqrt(N)*X[k]*np.exp(1j*2*np.pi*n*k/N)

# Loop de variâncias
error = np.zeros(len(vtEbN0_dB))
BER_simulada = np.zeros(len(vtEbN0_dB))
for ik in range(len(vtEbN0_dB)):
    # Adição de ruído
    noise = 1/np.sqrt(2)*(np.random.randn(int(N))+1j*np.random.randn(int(N)))
    noise = noise*10**(-vtEbN0_dB[ik]/20)
    # sinal recebido = xn + ruído 
    rn = xn+noise
    # DFT de rn
    Y = np.zeros((K), dtype=complex)
    #Y = (1/np.sqrt(N))*np.fft.fft(rn)
    Y = np.fft.fft(rn, norm='ortho')
    #for k in range(K):
    #    for n in range(N):
    #        Y[k] = Y[k] + 1/np.sqrt(N)*rn[n]*np.exp(-1j*2*np.pi*k*n/N)
    
    # Demodulação 
    Z = np.zeros(len(Y), dtype=complex)
    for k in range(len(Y)): # Para percorrer todo o vetor Yk 
        if np.real(Y[k]) > 0: # Para parte real de Yk positiva
            if np.real(Y[k]) > 2:
                Z[k] = 3
            else:
                Z[k] = 1
        else: # Para parte real de Yk negativa ou igual a zero
            if np.real(Y[k]) < -2:
                 Z[k] = -3
            else:
                 Z[k] = -1

        if np.imag(Y[k]) > 0: # Para parte imaginaria de Yk positiva
            if np.imag(Y[k]) > 2:
                Z[k] = Z[k] + 1j*3
            else:
                Z[k] = Z[k] + 1j

        else: # Para parte imaginaria de Yk negativa ou igual a zero
            if np.imag(Y[k]) < -2:
                 Z[k] = Z[k] - 1j*3
            else:
                 Z[k] = Z[k] - 1j

    # Contagem de erro
    error= len(np.where(Z-X!=0)[0])
    print("Para EbN0 = " + str(vtEbN0_dB[ik])  + " dB, ha " + str(error) + " simbolos errados" )
    
    #BER simulada
    BER_simulada[ik]=error/(n_bits/4)
    
# BER teórica
BER_teorica = (3/(2*4))*erfc(np.sqrt((4/10)*vtEbN0))

plt.figure(1,[14,7])
# AWGN simulado
plt.semilogy(vtEbN0_dB,BER_simulada,'ro',label='AWGN simulada')
# AWGN teórico
plt.semilogy(vtEbN0_dB,BER_teorica,'b--',label='AWGN teórica')
plt.grid()
plt.legend()
plt.title("E_b/N_0 Vs BER para 16QAM")
plt.xlabel('E_b/N_0(dB)')
plt.ylabel('BER ou Pe')
plt.show()
```
