```matlab
import numpy as np
import matplotlib.pyplot as plt 
#
# Parmetros
N = 20000                                  # Nmero de amostras
ts = 0.1
x=np.arange(0,5+ts,ts)                     # Eixo x
sigma = 1                                  # Parmetro Rayleigh
u = np.random.rand(N)                      # Amostras aleatrias uniformemente distribudas de 0 a 1
#
# Canal Rayleigh real via mtodo da inverso
rReal = sigma*np.sqrt(-2*np.log(u))        # Rayleigh via mtodo da inverso
#
# PDF Rayleigh terico 
pdfTeo = x/(sigma**2)*np.exp(-(x/sigma)**2/2) 
# 
# Canal Rayleigh complexo via VAs Gaussianas independentes
rComplexo = np.random.randn(N)+1j*np.random.randn(N)
#
# Grficos
plt.figure(1,[10,7])
plt.subplot(3,1,1)
# Histograma do Canal real vs PDF terica
[xh, temp]=np.histogram(rReal,bins=x)
temp = temp[0:len(temp)-1]
width = ts
plt.bar(temp,xh/(np.sum(xh)*ts),width,label='Histrograma das amostras - canal real')
#
# PDF terica
plt.plot(x,pdfTeo,'r', label='PDF terica Rayleigh')
#
plt.title('PDF da envoltria do canal Rayleigh real')
plt.legend()
plt.show()
#
# Envoltria do Canal complexo
plt.figure(1,[10,7])
plt.subplot(3,1,2)
# Histograma da envoltria do canal complexo vs PDF terica
[xh, temp]=np.histogram(np.abs(rComplexo),bins=x)
temp = temp[0:len(temp)-1]
width = ts
plt.bar(temp,xh/(np.sum(xh)*ts),width,label='Histrograma das amostras - canal complexo')
#
# PDF terica
plt.plot(x,pdfTeo,'r', label='PDF terica Rayleigh')
#
plt.title('PDF da envoltria do canal Rayleigh Complexo')
plt.legend()
plt.show()
# Histrograma da Fase do Canal complexo
plt.figure(1,[10,7])
plt.subplot(3,1,3)
[xhra, tempra]=np.histogram(np.angle(rComplexo))
tempra = tempra[0:len(tempra)-1]
plt.bar(180/np.pi*tempra,xhra/(np.sum(xhra)*ts),width=30)
plt.title('PDF da fase do canal Rayleigh Complexo')
plt.show()
```
