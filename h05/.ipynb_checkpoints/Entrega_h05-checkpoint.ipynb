## Prática 1: Modulação AM-DSB

### Parte 1 - expressando três casos clássicos de modulação

#### i) Modulação a 100%

Para que haja correto funcionamento de um transmissor de AM de modo que seja obtido maior alcance e para que ele não cause interferências, a modulação deve ser feita de forma bem determinada.

Na prática, a modulação é medida na forma de uma porcentagem. Assim, uma modulação de 100% ($M_u = 1$) é aquela em que a amplitude do sinal varia entre zero e o valor máximo.

```python
# Importa bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

# Parâmetros do sinal
Ac = 2                                                         # Amplitude da portadora
Mu = 1.0                                                       # Índice de modulação
fc = 25000                                                     # Frequência da portadora Hz
fm = 2000
N = 1000
Ts = 1e-6                                                      # Tempo de amostragem pequeno (modelar sinal contínuo)
t = np.arange(N)*Ts
s = Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)

# Cálculo da FFT de AM-DSB
lfft = 30                                                      # Número pontos da fft
k = np.arange(-lfft,lfft)                                      # Vetor de frequências 
S_f = 2.0*np.abs((fftpack.fft(s)))/N                           # Cálculo da FFT
Ns = len(s)                                                    # Comprimento do sinal modulado
Nk = len(k)                                                    # Comprimento do sinal em frequência

# A fft em 30 pontos (para melhor visualização)
S_f_new = np.zeros(Nk)                                         # Inicialização do vetor da frequência
fsampling = 1/Ts                                               # Taxa de amostragem
freq = (fsampling/Ns)*k                                        # Eixo de frequências
for i in range(Nk):
    kk = k[i]
    if kk>=0:
        S_f_new[i] = S_f[kk]
    else :
        S_f_new[i] = S_f[Ns+kk]

# Gráfico do AM-DSB no tempo
plt.figure(1,[10,7])
plt.subplot(211)
plt.plot(t,s)
plt.title("Sinal AM no tempo - AM-DSB (padrão)")
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")

# Gráfico do AM-DSB na frequência
plt.subplot(212)
plt.title("AM-DSB na frequência")
plt.xlabel("Frequência [kHz]")
plt.ylabel("Magnitude")
plt.stem(freq/1e3,S_f_new)
plt.tight_layout()
plt.show()
```

#### ii) Submodulação

Uma modulação pobre, menor que 100%, em que a amplitude da portadora varie pouco de intensidade, faz com que o rendimento seja baixo e, além disso, não faz com que o sinal varie em toda sua amplitude, dificultando a identificação da envoltoria.

```python
# Importa bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

# Parâmetros do sinal
Ac = 2                                                         # Amplitude da portadora
Mu = 0.3                                                       # Índice de modulação
fc = 25000                                                     # Frequência da portadora Hz
fm = 2000
N = 1000
Ts = 1e-6                                                      # Tempo de amostragem pequeno (modelar sinal contínuo)
t = np.arange(N)*Ts
s = Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)

# Cálculo da FFT de AM-DSB
lfft = 30                                                      # Número pontos da fft
k = np.arange(-lfft,lfft)                                      # Vetor de frequências 
S_f = 2.0*np.abs((fftpack.fft(s)))/N                           # Cálculo da FFT
Ns = len(s)                                                    # Comprimento do sinal modulado
Nk = len(k)                                                    # Comprimento do sinal em frequência

# A fft em 30 pontos (para melhor visualização)
S_f_new = np.zeros(Nk)                                         # Inicialização do vetor da frequência
fsampling = 1/Ts                                               # Taxa de amostragem
freq = (fsampling/Ns)*k                                        # Eixo de frequências
for i in range(Nk):
    kk = k[i]
    if kk>=0:
        S_f_new[i] = S_f[kk]
    else :
        S_f_new[i] = S_f[Ns+kk]

# Gráfico do AM-DSB no tempo
plt.figure(1,[10,7])
plt.subplot(211)
plt.plot(t,s)
plt.title("Sinal AM no tempo - AM-DSB (padrão)")
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")

# Gráfico do AM-DSB na frequência
plt.subplot(212)
plt.title("AM-DSB na frequência")
plt.xlabel("Frequência [kHz]")
plt.ylabel("Magnitude")
plt.stem(freq/1e3,S_f_new)
plt.tight_layout()
plt.show()
```

#### iii) Sobremodulação
Pode ocorrer, ainda, que a modulação supere os 100%, caso em que teremos instantes em que o sinal se anula. Essa modulação excessiva, que ultrapassa a variação da amplitude do sinal de alta frequência, faz com que ocorram distorções e sejam geradas interferências.

Um sinal não deve ser modulado em mais do que 100%, pois além de ocorrerem distorções dos sinais, são geradas bandas laterais fora das faixas designadas causando interferências. Neste caso, dizemos que ocorre uma sobremodulação.

```python
# Importa bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

# Parâmetros do sinal
Ac = 2                                                         # Amplitude da portadora
Mu = 1.5                                                       # Índice de modulação
fc = 25000                                                     # Frequência da portadora Hz
fm = 2000
N = 1000
Ts = 1e-6                                                      # Tempo de amostragem pequeno (modelar sinal contínuo)
t = np.arange(N)*Ts
s = Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)

# Cálculo da FFT de AM-DSB
lfft = 30                                                      # Número pontos da fft
k = np.arange(-lfft,lfft)                                      # Vetor de frequências 
S_f = 2.0*np.abs((fftpack.fft(s)))/N                           # Cálculo da FFT
Ns = len(s)                                                    # Comprimento do sinal modulado
Nk = len(k)                                                    # Comprimento do sinal em frequência

# A fft em 30 pontos (para melhor visualização)
S_f_new = np.zeros(Nk)                                         # Inicialização do vetor da frequência
fsampling = 1/Ts                                               # Taxa de amostragem
freq = (fsampling/Ns)*k                                        # Eixo de frequências
for i in range(Nk):
    kk = k[i]
    if kk>=0:
        S_f_new[i] = S_f[kk]
    else :
        S_f_new[i] = S_f[Ns+kk]

# Gráfico do AM-DSB no tempo
plt.figure(1,[10,7])
plt.subplot(211)
plt.plot(t,s)
plt.title("Sinal AM no tempo - AM-DSB (padrão)")
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")

# Gráfico do AM-DSB na frequência
plt.subplot(212)
plt.title("AM-DSB na frequência")
plt.xlabel("Frequência [kHz]")
plt.ylabel("Magnitude")
plt.stem(freq/1e3,S_f_new)
plt.tight_layout()
plt.show()
```

## Prática 2: recepção AM-DSB (detecção de envoltória)

### Para $\tau=10^{-4}$

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import fftpack

tau = 1e-4                                                      # Constante de tempo do detector de envelope
Ts=1e-6                                                         # Definição do período
t = np.arange(1000)*Ts                                          # Definição do vetor tempo
fc = 10000                                                      # Frequência da portadora.
fm = 1500                                                       # Frequência do sinal
Mu = 0.6                                                        # Índice de modulaçao.
Ac = 1.0
x_AMo = Ac*(1.0+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t);  # Onda Modulada AM-DSB

x_envIdeal = np.abs(Ac*(1.0+Mu*np.cos(2*np.pi*fm*t)))           # Envoltória ideal

# Detector de envoltória
x_AM = x_AMo*(x_AMo>0)                                          # Efeito do diodo (semiciclo positivo)
x_env = np.zeros(len(x_AM))
Ns = len(x_AM)
out = -1
for i in range(Ns):
    inp = x_AM[i]
    if inp>=out:
        out = inp            # Caso 1: x_am(t) > Vc(t) (carga do capacitor)
    else:
        out *= (1-Ts/tau)    # Caso 2: x_am(t) < Vc(t) (descarga do capacitor)
    x_env[i] = out

print('Erro médio quadrático: ', sum((x_env-x_envIdeal)**2)/len(x_env))

# gráfico composto
plt.figure(1,[10,7])
plt.title("Detecção de envoltória")
plt.ylabel("Amplitude")
plt.xlabel("Tempo(ms)")
envoltoria_ideal = plt.plot(t*1000,x_envIdeal)
sinal_transmitido = plt.plot(t*1000,x_AM)
detector_de_saida = plt.plot(t*1000,x_env)
plt.grid()
plt.legend(["Envoltória ideal","Onda AM retificada","Envoltória recuperada"])
plt.show()
```

### Para $\tau = 4\cdot10^{-3}$

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import fftpack

tau = 4*1e-3                                                      # Constante de tempo do detector de envelope
Ts=1e-6                                                         # Definição do período
t = np.arange(1000)*Ts                                          # Definição do vetor tempo
fc = 10000                                                      # Frequência da portadora.
fm = 1500                                                       # Frequência do sinal
Mu = 0.6                                                        # Índice de modulaçao.
Ac = 1.0
x_AMo = Ac*(1.0+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t);  # Onda Modulada AM-DSB

x_envIdeal = np.abs(Ac*(1.0+Mu*np.cos(2*np.pi*fm*t)))           # Envoltória ideal

# Detector de envoltória
x_AM = x_AMo*(x_AMo>0)                                          # Efeito do diodo (semiciclo positivo)
x_env = np.zeros(len(x_AM))
Ns = len(x_AM)
out = -1
for i in range(Ns):
    inp = x_AM[i]
    if inp>=out:
        out = inp            # Caso 1: x_am(t) > Vc(t) (carga do capacitor)
    else:
        out *= (1-Ts/tau)    # Caso 2: x_am(t) < Vc(t) (descarga do capacitor)
    x_env[i] = out

print('Erro médio quadrático:', sum((x_env-x_envIdeal)**2)/len(x_env))

# gráfico composto
plt.figure(1,[10,7])
plt.title("Detecção de envoltória")
plt.ylabel("Amplitude")
plt.xlabel("Tempo(ms)")
envoltoria_ideal = plt.plot(t*1000,x_envIdeal)
sinal_transmitido = plt.plot(t*1000,x_AM)
detector_de_saida = plt.plot(t*1000,x_env)
plt.grid()
plt.legend(["Envoltória ideal","Onda AM retificada","Envoltória recuperada"])
plt.show()
```

Como visto nas simulações e nos gráficos plotados, dependendo da constante de descarga do capacitor, a detecção (e o erro quadrático médio) será maior ou menor). Caso $\tau$ seja muito menor que o período da portadora, o capacitor se descarregará muito rápido quando a portadora cai abaixo do seu valor de pico, o que fará com que a saída do detector não siga a envoltória desta onda.

Se, por outro lado, o valor for muito maior que a máxima do sinal modulador, então o capacitor se descarregará muito lentamente e a saída do detector não seguirá a envoltória da onda AM, resultando em um corte diagonal do sinal.

## Prática 8: Sinal analítico, Transformada de Hilbert e seu uso em demodulação

```python
# Sinal modulante:
 # Frequência de amostragem = 600;
 # Eixo do tempo de 0 a 1;
 # Frequência do sinal = 3;
# Sinal portadora:
 # Frequência da portadora = 50;
 # Indice de modulação = 0,7

%matplotlib inline

# Importa bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
from scipy.signal import hilbert

# Parâmetros do sinal
Ac = 1                                        # Amplitude da portadora
Mu = 0.7                                      # Índice de modulação
fc = 50                                       # Frequência da portadora Hz
fm = 3
N = 1000
Ts = 1/600                                    # Tempo de amostragem pequeno (modelar sinal contínuo)
t = np.arange(N)*Ts
m_t = np.cos(2*np.pi*fm*t)
s = Ac*(1+Mu*m_t)*np.cos(2*np.pi*fc*t)        # Sinal modulado

# Hilbert
z_t = hilbert(s)
m_est = abs(z_t)
ang = np.angle(z_t)
portadora = np.cos(ang)


def fft(signal):
    # Cálculo da FFT modulante
    lfft = 30                                         # Número pontos da fft
    k = np.arange(-lfft,lfft)                         # Vetor de frequências 
    S_f = 2.0*np.abs((fftpack.fft(signal)))/N         # Cálculo da FFT
    Ns = len(signal)                                  # Comprimento do sinal modulado
    Nk = len(k)                                       # Comprimento do sinal em frequência

    # A fft em 30 pontos (para melhor visualização)
    fsampling = 1/Ts
    freq = (fsampling/Ns)*k
    
    fft_signal = np.zeros(Nk)                         # Inicialização do vetor da frequência
    for i in range(Nk):
        kk = k[i]
        if kk>=0:
            fft_signal[i] = S_f[kk]
        else:
            fft_signal[i] = S_f[Ns+kk]
    return fft_signal, freq


[fft_modulante, freq] = fft(m_t)
[fft_modulado, freq2] = fft(s-1)
[fft_regenerado, freq3] = fft(m_est)

############ Graficos ################
# Gráfico do sinal modulante, do sinal modulado e do sinal regenerado no tempo
plt.title('Modulação e demodulação AM pela transformada de Hilbert')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.plot(t,m_t,t,s,t,m_est)
plt.ylim([-2,2])
plt.xlim([0,1])
plt.legend(['sinal modulante','sinal modulado', 'sinal regenerado', 'angle'])
plt.show()

# Espectro do sinal modulante e regenerado
plt.plot(freq/1e3,fft_modulante, freq3/1e3,fft_regenerado)
plt.title('Espectro do sinal modulante e regenerado')
plt.xlabel('Frequência [kHz]') 
plt.ylabel('Magnitude')
plt.legend(['sinal modulante','sinal regenerado'])
plt.tight_layout()
plt.show()

# Espectro do sinal modulado
plt.plot(freq2/1e3,fft_modulado)
plt.title('Espectro do sinal modulado')
plt.xlabel('Frequência [kHz]') 
plt.ylabel('Magnitude')
plt.legend(['sinal modulado'])
plt.tight_layout()
plt.show()

# Portadora regenerada por meio da transformada de Hilbert
plt.plot(t, portadora, t, np.cos(2*np.pi*fc*t))
plt.title('Portadora regenerada por meio da transformada de Hilbert')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.ylim([-2,2])
plt.xlim([0,1])
plt.legend(['portadora regenerada', 'portadora'])
plt.show()
```
