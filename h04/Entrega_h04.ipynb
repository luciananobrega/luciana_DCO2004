## Função *periodogram*

```matlab
 % Calcular o periodograma de dois lados de uma senoide de 200 Hz com ruído usando a janela Hamming.
       Fs = 1000;   t = 0:1/Fs:.3;
       x = cos(2*pi*t*200)+randn(size(t));
       plot(x,t)
       periodogram(x,[],'twosided',512,Fs);
```

- Frequencia de 200 Hz, centralizado em 500 (Fs/2).

-- When Fs is specified, the interval becomes [0,Fs).

-- maior potencia em 200 Hz. por isso fica em 200, mesmo com o ruido
