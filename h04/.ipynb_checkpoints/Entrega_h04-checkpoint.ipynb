## Função *periodogram*

```matlab
 % Calcular o periodograma de dois lados de uma senoide de 200 Hz com ruído usando a janela Hamming.
       Fs = 1000;   t = 0:1/Fs:.3;
       x = cos(2*pi*t*200)+randn(size(t));
       plot(x,t)
       periodogram(x,[],'twosided',512,Fs);
```

O sinal possui frequencia de 200 Hz e, portanto, com maior relacao potencia/frequencia, apesar do ruido. A frequencia central eh de Fs/2, com intervalo [0, Fs).

Como eh um periodograma de dois lados, os valores apresentados apos 500 Hz sao valores espelhados das frequencias menores que a frequencia central.

Analisando o mesmo sinal, sem ruido, percebe-se mais claramente as frequencias de maior relacao dB/Hz.

```matlab
 % Calcular o periodograma de dois lados de uma senoide de 200 Hz sem ruído usando a janela Hamming.
       Fs = 1000;   t = 0:1/Fs:.3;
       x = cos(2*pi*t*200);
       plot(x,t)
       periodogram(x,[],'twosided',512,Fs);
```

Abaixo, eh analisado o periodograma de um lado desse sinal sem ruido.

```matlab
 % Calcular o periodograma de um lado (one sided) de uma senoide de 200 Hz sem ruído usando a janela Hamming.
       Fs = 1000;   t = 0:1/Fs:.3;
       x = cos(2*pi*t*200);
       plot(x,t)
       periodogram(x,[],'onesided',512,Fs);
```
