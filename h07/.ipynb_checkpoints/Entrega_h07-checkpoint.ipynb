### 1.1. Lognormal

#### a) Python

```python
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt

%matplotlib inline

# Parâmetros da distribuição
vtMu = [0, -10, 10]                  # Valores de média da Gaussiana
vtVar = [1, 5, 10]                   # Valores de variância da Gaussiana
x = np.arange(-20,20,0.1)     
#
# Variando a média e plotando os gráficos
plt.figure(1,[15,5])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    dist=lognorm(sigma,mu)
    plt.subplot(2,2,1)
    plt.plot(x, dist.pdf(x), label='Média = {}'.format(mu))
    plt.subplot(2,2,2)
    plt.plot(x, dist.cdf(x), label='Média = {}'.format(mu))
    
# Variando a variância e plotando os gráficos
mu = vtMu[0];
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    dist=lognorm(sigma,loc=mu)
    plt.subplot(2,2,3)
    plt.plot(x, dist.pdf(x), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.subplot(2,2,4)
    plt.plot(x, dist.cdf(x), label='$\sigma$ = {:01.2f}'.format(sigma))
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()
```

### 1.2. Rayleigh

#### a) Python

```python
import numpy as np
from scipy.stats import rayleigh
import matplotlib.pyplot as plt

%matplotlib inline

# Parâmetros da distribuição
vtMu = [0, -10, 10]                  # Valores de média da Gaussiana
vtVar = [1, 5, 10]                   # Valores de variância da Gaussiana
x = np.arange(-20,20,0.1)     
#
# Variando a média e plotando os gráficos
plt.figure(1,[15,5])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    dist=rayleigh(x,mu)
    plt.subplot(2,2,1)
    plt.plot(x, rayleigh.pdf(x,mu), label='Média = {}'.format(mu))
    plt.subplot(2,2,2)
    plt.plot(x, rayleigh.cdf(x,mu), label='Média = {}'.format(mu))
    
# Variando a variância e plotando os gráficos
mu = vtMu[0];
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, rayleigh.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.subplot(2,2,4)
    plt.plot(x, rayleigh.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()
```

A duração de chamadas em uma rede telefônica é uma VA com distribuição exponencial de taxa λ, em chamadas/segundo. Duas operadoras mediram: λ1=2 e λ2=5. Foi requisitado que você calculasse qual a probabilidade de assinantes fazerem chamadas com mais de 3 minutos nas duas operadoras. Faça um código que calcule e mostre graficamente as probabilidades. Faça seu código em Python.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sympy import *

# Cálculo da probabilidade
def calc_prob(sigma):
    mu = 0                                       # Média
    T=0.001                                      # Taxa de amostragem
    x=np.arange(0,5+T,T)                        # Eixo x
    DistNorm=stats.expon.pdf(x,mu,sigma)         # Distribuição exponencial
    
    ind = x.tolist().index(3)
    indices = np.arange(int(ind)+1, len(x))
    prob1=np.sum(DistNorm[indices])*T*100        # Probabilidade de um evento ocorrer no intervalo
    plt.plot(x,DistNorm,'k')                                       
    plt.title('Probabilidade de = ' + str(prob1) +'% para lambda = ' + str(1/sigma))      # Mostra valor verdadeiro de prob1
    plt.fill_between(x[indices],DistNorm[indices],facecolor='midnightblue')
    #plt.ylim([0,.00002])
    plt.show()


# Distribuição

lamb1 = 2; sigma1 = 1/lamb1
lamb2 = 5; sigma2 = 1/lamb2

calc_prob(sigma1)
calc_prob(sigma2)
```
