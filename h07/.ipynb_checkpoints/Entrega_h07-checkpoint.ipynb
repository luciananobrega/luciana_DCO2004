### 1.1. Lognormal
Como é possível observar nos gráficos abaixo, para um mesmo desvio padrão, a média indica, aproximadamente, onde a PDF e CDF se iniciam, visto que a maior parte dos valores gerados por uma distribuição log-normal deve próxima (mas um pouco acima) do seu limite inferior. O padrão de queda da PDF também é a mesma para um mesmo desvio padrão.

O desvio padrão determina o formato da PDF, e isso pode ser observado no gráfico de PDFs em que a média é constante.
Por ter a média constante, as curvas dos gráficos de CDF de diferentes desvios padrões se encontram em 50%.

```python
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt

%matplotlib inline

# Parâmetros da distribuição
vtMu = [0, -10, 10]                  # Valores de média da Gaussiana
vtVar = [1, 5, 20]                   # Valores de variância da Gaussiana
x = np.arange(-20,20,0.1)     
#
# Variando a média e plotando os gráficos
plt.figure(1,[15,5])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    dist=lognorm(sigma,mu)
    plt.subplot(2,2,1)
    plt.plot(x, dist.pdf(x), label='Média = {}'.format(mu))
    plt.subplot(2,2,2)
    plt.plot(x, dist.cdf(x), label='Média = {}'.format(mu))
    
# Variando a variância e plotando os gráficos
mu = vtMu[0];
x = np.arange(0,5,0.1) 
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    dist=lognorm(sigma,loc=mu)
    plt.subplot(2,2,3)
    plt.plot(x, dist.pdf(x), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.subplot(2,2,4)
    plt.plot(x, dist.cdf(x), label='$\sigma$ = {:01.2f}'.format(sigma))
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()
```

### 1.2. Rayleigh

Para um mesmo desvio padrão e médias diversas, o comportamento da distribuição de Rayleigh é similar ao do Lognormal.

Porém, para uma mesma média e diferentes desvios padrões, observa-se que os valores são mais dispersos com o aumento do desvio padrão (PDF mais larga). A CDF, por consequência, é mais inclinada e demora mais para atingir o ponto de 100%.

```python
import numpy as np
from scipy.stats import rayleigh
import matplotlib.pyplot as plt

%matplotlib inline

# Parâmetros da distribuição
vtMu = [0, -10, 10]            # Valores de média
vtVar = [1, 5, 10]             # Valores de variância
x = np.arange(-20,20,0.1)     
#
# Variando a média e plotando os gráficos
plt.figure(1,[15,5])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    dist=rayleigh(x,mu)
    plt.subplot(2,2,1)
    plt.plot(x, rayleigh.pdf(x,mu), label='Média = {}'.format(mu))
    plt.subplot(2,2,2)
    plt.plot(x, rayleigh.cdf(x,mu), label='Média = {}'.format(mu))
    
# Variando a variância e plotando os gráficos
mu = vtMu[0];
x = np.arange(0,20,0.1)  
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, rayleigh.pdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.subplot(2,2,4)
    plt.plot(x, rayleigh.cdf(x,mu,sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()
```

### 1.3 Rice

Para um mesmo desvio padrão e médias diversas, o comportamento da distribuição de Rice é similar ao do Lognormal e Rayleigh. Porém, para um mesmo desvio padrão, a distribuição de Rayleigh apresenta valores mais diversos (PDF mais larga) quando comparada às duas distribuições anteriores. Além disso, quanot maior a média, mais próxima a distribuição se torna de uma distribuição gaussiana.
Porém, para uma mesma média e diferentes desvios padrões, observa-se que os valores são mais dispersos com o aumento do desvio padrão (PDF mais larga). A CDF, por consequência, é mais inclinada e demora mais para atingir o ponto de 100%.



```python
import numpy as np
from scipy.stats import rice
import matplotlib.pyplot as plt
#
# Parâmetros da distribuição
vtMu = [0, 5, 10]               # Valores de média
vtVar = [1, 5, 10]              # Valores de variância
x = np.arange(0,20,0.1)     
#
# Variando a média e plotando os gráficos
plt.figure(1,[15,5])
sigma = np.sqrt(vtVar[0]);
for il in range(0,len(vtMu)):
    mu = vtMu[il]
    plt.subplot(2,2,1)
    plt.plot(x, rice.pdf(x,b=mu, scale=sigma), label='Média = {}'.format(mu))
    plt.subplot(2,2,2)
    plt.plot(x, rice.cdf(x,b=mu, scale=sigma), label='Média = {}'.format(mu))  
    
# Variando a variância e plotando os gráficos
mu = vtMu[0]
for il in range(0,len(vtVar)):
    sigma = np.sqrt(vtVar[il]);
    plt.subplot(2,2,3)
    plt.plot(x, rice.pdf(x,b=mu,scale=sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
    plt.subplot(2,2,4)
    plt.plot(x, rice.cdf(x,b=mu,scale=sigma), label='$\sigma$ = {:01.2f}'.format(sigma))
        
plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()
```

### 1.4. Nakagami

Na distribuição de Nakagami, o parâmetro m determina o formato da curva da PDF, a qual se torna menos larga e mais alta com a redução de m. Já o parâmetro $\Omega$ determina o espalhamento da curva de PDF. A curva se torna mais dispersa com o seu aumento. Na CDF, o parâmetro torna a curva mais ou menos inclinada (de maneira inversamente proporcional), de forma que, quanto maior o valor de $\Omega$, mais demorado é para a curva atingir o ponto de 100%.

```python
import numpy as np
from scipy.stats import nakagami
import matplotlib.pyplot as plt

# Parâmetros da distribuição
vtMu = [1,2,4]
vtOmega = [1, 2, 4]
x = np.arange(0,10,0.01)     


plt.figure(1,[15,5])
for il in range(0,len(vtMu)):
    mu=vtMu[il]
    plt.subplot(2,2,1)
    plt.plot(x, nakagami.pdf(x,nu=mu,scale=vtOmega[0]), label='Média = {}'.format(mu))
    plt.subplot(2,2,2)
    plt.plot(x, nakagami.cdf(x,nu=mu,scale=vtOmega[0]), label='Média = {}'.format(mu))
    
mu = vtMu[0];   
for il in range(0,len(vtOmega)):
    plt.subplot(2,2,3)
    plt.plot(x, nakagami.pdf(x,nu=mu,scale=vtOmega[il]), label='$\Omega$ = {:01.2f}'.format(vtOmega[il]))
    plt.subplot(2,2,4)
    plt.plot(x, nakagami.cdf(x,nu=mu,scale=vtOmega[il]), label='$\Omega$ = {:01.2f}'.format(vtOmega[il]))

plt.subplot(2,2,1)
plt.legend()
plt.subplot(2,2,2)
plt.legend()
plt.subplot(2,2,3)
plt.legend()
plt.subplot(2,2,4)
plt.legend()
plt.show()
```

### 1.6 Exponencial
Na distribuição exponencial, o parâmetro $\lambda$ (eventos/segundo) indica a probabilidade de um evento continuar ocorrendo após determinado período. Quanto maior o parâmetro $\lambda$, mais amostras finalizam mais cedo e a curva decai mais rapidamente, e isso é evidenciado na PDF.

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt

# Parâmetros da distribuição
lamb = [1,2,5]
x = np.arange(0,2,0.01)     

# Variando lambda e plotando os gráficos
plt.figure(1,[15,5])
for il in range(0,len(lamb)):
    plt.subplot(1,2,1)
    plt.plot(x, expon.pdf(x,scale=1/lamb[il]), label='$\lambda$ = {}'.format(lamb[il]))
    plt.subplot(1,2,2)
    plt.plot(x, expon.cdf(x,scale=1/lamb[il]), label='$\lambda$ = {}'.format(lamb[il]))

plt.subplot(1,2,1)
plt.legend()
plt.subplot(1,2,2)
plt.legend()
plt.show()
```

### 1. Turbina de um avião
A vida efetiva de um componente de turbina de avião é de 5000h e desvio padrão de 40h. A distribuição da vida efetiva é bastante próxima da distribuição normal. Faça um código que calcule e mostre graficamente a probabilidade de que uma determinada amostra tenha vida efetiva maior que 5038h. Faça seu código em Matlab.

```python
clc;clear all; close all;
%% Parâmetros
mu = 5000;                                              % Média
sigma = 40;                                             % Desvio padrâo
T=0.001;                                                % Taxa de amostragem
x=4700:T:5300;                                          % Eixo x
v_eff = 5038;                                           % Vida efetiva procurada

% Distribuição
DistNorm=normpdf(x,mu,sigma);                           % Distribuição normal    
% Estimação da probabilidade
%indices = (max(find(x<(mu-sigma))):min(find(x>(mu+sigma))));     % Encontra o local que se deseja estimar
indices = (max(find(x==v_eff)):length(x));
prob1=sum(DistNorm(indices))*T*100;                     % Probabilidade de um evento ocorrer no intervalo
plot(x,DistNorm);                                       
title([ 'Probabilidade de =  ' num2str(prob1)  ]);      % Mostra valor verdadeiro de prob1
hold on;
area(x(indices),DistNorm(indices));                      % Probabilidade 

disp([' Probabilidade pela integral da fórmula da PDF = ' num2str(prob1) ' %']);
```

### 2. Duração de chamadas
A duração de chamadas em uma rede telefônica é uma VA com distribuição exponencial de taxa λ, em chamadas/segundo. Duas operadoras mediram: λ1=2 e λ2=5. Foi requisitado que você calculasse qual a probabilidade de assinantes fazerem chamadas com mais de 3 minutos nas duas operadoras. Faça um código que calcule e mostre graficamente as probabilidades. Faça seu código em Python.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sympy import *

# Cálculo da probabilidade
def calc_prob(sigma):
    mu = 0                                       # Média
    T=0.001                                      # Taxa de amostragem
    x=np.arange(0,5+T,T)                        # Eixo x
    DistExp=stats.expon.pdf(x,mu,sigma)         # Distribuição exponencial
    
    ind = x.tolist().index(3)
    indices = np.arange(int(ind)+1, len(x))
    prob1=np.sum(DistExp[indices])*T*100        # Probabilidade de um evento ocorrer no intervalo
    plt.plot(x,DistExp,'k')                                       
    plt.title('Probabilidade de = ' + str(prob1) +'% para lambda = ' + str(1/sigma))      # Mostra valor verdadeiro de prob1
    plt.fill_between(x[indices],DistExp[indices],facecolor='midnightblue')
    plt.show()


# Distribuição

lamb1 = 2; sigma1 = 1/lamb1
lamb2 = 5; sigma2 = 1/lamb2

calc_prob(sigma1)
calc_prob(sigma2)
```

## Prática 02: Geração de variáveis aleatórias com uma PDF específica

Nesta prática, serão geradas amostras da VA que representa a observação do número que aparece na face de um dado não tendencioso.

Percebe-se, pelos gráficos gerados, que a PDF estimada fica cada vez mais próxima da teórica à medida que o número de amostras aumenta.

```python
# Geração de Variável aleatória
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import plotly.plotly as py

# Parâmetros da Gaussiana
vtnSamples = [1e2, 1e3, 1e5] # Número de amostras

# Estimativa da PDF (PMF)
x = np.arange(1,7,1)
a = randint.pmf(x, 1,7) 

for i in range(len(vtnSamples)):
    # Gerar números aleatórios entre 1 e 6 (incluso):
    dado = np.random.randint(1, 7, int(vtnSamples[i]))
    
    # Plot
    plt.figure(i)
    plt.bar(x,a, color='red')
    plt.hist(dado, np.arange(1,8,1), normed=True, align='left', rwidth=0.6)
    plt.ylabel('Probabilidade')
    plt.title('Para ' + str(int(vtnSamples[i])) + ' amostras')
    plt.legend(['Teórico', 'Estimada'])
    plt.show()
```
